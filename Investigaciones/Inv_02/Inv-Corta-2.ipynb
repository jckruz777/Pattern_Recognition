# Métodos de Aprendizaje Supervisado

## Introducción

En este trabajo se presenta la exploración de 6 métodos de Aprendizaje Automático Supervisado, aplicados sobre un set de datos denominado [Global Shark Attack File](https://toolbox.google.com/datasetsearch/search?query=Shark%20attacks&docid=3FpN5ndbypMlKYa3AAAAAA%3D%3D). 

### Descripción del Set
El set de datos seleccionado presenta información sobre los ataques de tiburón a humanos registrados desde el siglo XV. Los atributos considerados de mayor relevancia de este set de datos corresponden a los siguientes:

* __Año:__ año del ataque registrado.
* __País:__ país donde se registró el ataque.
* __Actividad:__ actividad realizada por la víctima previo al ataque.
* __Sexo:__ sexo de la víctima.
* __Edad:__ edad de la víctima.
* __Fatalidad:__ indicador de si el ataque fue fatal o no.

### Objetivo
Evaluar 6 métodos de Aprendizaje Supervisado en la estimación de la fatalidad de ataques de tiburón a humanos, considerando la actividad realizada previo al ataque, tanto para mujeres como para hombres entre los 10 y los 75 años.

En la siguiente sección de código se cargan los datos del set y se filtran por los atributos de interés:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

read_df = pd.read_excel("GSAF5.xls")
interest_df = read_df.filter(items=['Year', 'Country', 'Activity', 'Sex', 'Age', 'Fatal']) 
interest_df.head(20)
```

Ahora se filtran los datos para considerar únicamente:
* Los casos donde todos los atributos sean conocidos.
* Los ataques registrados entre 1900 y el 2018.
* Los ataques donde la víctima tiene una edad entre los 10 y los 75 años
* Las siguientes 20 actividades:
  * "Body boarding"
  * "Body surfing"
  * "Diving"
  * "Feeding sharks"
  * "Freediving"
  * "Freedom swimming"
  * "Fishing"
  * "Kayaking"
  * "Night bathing"
  * "Photo shoot"
  * "Scuba diving"
  * "Shark fishing"
  * "Shipwreck"
  * "Snorkeling"
  * "Spearfishing"
  * "Standing"
  * "Surfing"
  * "Swimming"
  * "Wading"
  * "Walking"

```python
recent_df = interest_df[interest_df.Year >= 1900.0]
recent_df.Age = recent_df.Age.apply(pd.to_numeric, errors='coerce')
recent_df = recent_df.dropna()
fcountry_df = recent_df[recent_df.Country.str.len() <= 20]
ffatal_df = fcountry_df[(fcountry_df.Fatal == "Y") | (fcountry_df.Fatal == "N")]
factivity_df = ffatal_df[(ffatal_df.Activity == 'Body boarding')    |
                         (ffatal_df.Activity == 'Body surfing')     |
                         (ffatal_df.Activity == 'Diving')           |
                         (ffatal_df.Activity == 'Feeding sharks')   |
                         (ffatal_df.Activity == 'Freediving')       |
                         (ffatal_df.Activity == 'Freedom swimming') |
                         (ffatal_df.Activity == 'Fishing')          |
                         (ffatal_df.Activity == 'Kayaking')         |
                         (ffatal_df.Activity == 'Night bathing')    |
                         (ffatal_df.Activity == 'Photo shoot')      |
                         (ffatal_df.Activity == 'Scuba diving')     |
                         (ffatal_df.Activity == 'Shark fishing')    |
                         (ffatal_df.Activity == 'Shipwreck')        |
                         (ffatal_df.Activity == 'Snorkeling')       |
                         (ffatal_df.Activity == 'Spearfishing')     |
                         (ffatal_df.Activity == 'Standing')         |
                         (ffatal_df.Activity == 'Surfing')          |
                         (ffatal_df.Activity == 'Swimming')         |
                         (ffatal_df.Activity == 'Wading')           |
                         (ffatal_df.Activity == 'Walking')
                        ]
fage_df = factivity_df[(factivity_df.Age >= 10.0) & (factivity_df.Age <= 75.0)]
fage_df.head(20)
```

## División del Set: Entrenamiento y Prueba

Nótese que el set de datos está ordenados descendente utilizando el año como criterio. Dado esto, se procede a mezclar las filas del set para romper dicho orden:

```python
df = fage_df.sample(frac=1).reset_index(drop=True)
df.head(20)
```

Ahora se realiza la división. Después de haber aplicado los filtros, se cuenta con un set preprocesado de 1885 ataques registrados. Se ha tomado la decisión de separar el set de la siguiente manera: 
* __Entrenamiento:__ 80% del set pre-procesado (df)
* __Prueba:__ 20% del set pre-procesado (df)

```python
trn_percent = 0.80
tst_percent = 0.20

# Separacion por porcentajes
trn_size = int(trn_percent * len(df))
tst_size = int(len(df) - trn_size)

trn_set_df = df.head(trn_size)
tst_set_df = df.tail(tst_size)
trn_tgt_df  = trn_set_df.Fatal
tst_tgt_df  = tst_set_df.Fatal
trn_set_df = trn_set_df.drop("Fatal", axis=1)
tst_set_df = tst_set_df.drop("Fatal", axis=1)

print("TRAINING DATASET: " + str(trn_size) + " samples")
trn_set_df.head(20)
```

```python
print("TESTING DATASET:"  + str(tst_size) + " samples")
tst_set_df.head(20)
```

### Set de Entrenamiento

A continuación se muestra la distribución de los datos en un gráfico _Actividad vs Edad_ cuyas etiquetas están dadas por la fatalidad positiva y negativa: 

```python
# Extrayendo los datos y las clases
target = trn_tgt_df

# Creando objeto para las gráficas
_, ax = plt.subplots()

# Asignando colores por clase
colors = ("red", "blue")
groups = ("Fatal", "No Fatal")

# Iterando por cada clase
it = 0
for i in np.unique(target):
    mask = target == i
    plt.scatter(trn_set_df.Age[mask.values], trn_set_df.Activity[mask.values], color=colors[it], label=groups[it])
    it = it + 1
    
# Asignando títulos a los ejes
ax.set_title("Activity vs Age")
ax.set_xlabel("Age")
ax.set_ylabel("Activity")

# Agregando las etiquetas de los grupos
ax.legend()

# Mostrando la gráfica
plt.show()
```

Ahora se muestra la distribución de los datos en un gráfico _Actividad vs País_

```python
# Creando objeto para las gráficas
_, ax = plt.subplots()

# Iterando por cada clase
it = 0
for i in np.unique(target):
    mask = target == i
    plt.scatter(trn_set_df.Country[mask.values], trn_set_df.Activity[mask.values], color=colors[it], label=groups[it])
    it = it + 1

plt.xticks(rotation=90)

# Asignando títulos a los ejes
ax.set_title("Activity vs Country")
ax.set_xlabel("Country")
ax.set_ylabel("Activity")

# Agregando las etiquetas de los grupos
ax.legend()

# Mostrando la gráfica
plt.show()
```

### Encodificando Strings

Se procede a encodificar los valores de atributo correspondientes a strings para evitar errores a la hora de utilizar los métodos de aprendizaje supervisado. 

```python
from sklearn.preprocessing import LabelEncoder

def encode_dataset(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column] = le.fit_transform(dataset[column])
```

Se aplica a los sets de entrenamiento y de prueba:

```python
encode_dataset(trn_set_df)
encode_dataset(tst_set_df)
```

## Aprendizaje Supervisado

En esta sección se presentan 6 métodos de Aprendizaje Supervisado utilizados para predecir la fatalidad de los casos. Se pretende evaluar la precisión de cada método cuando éstos logran generalizar del set de entrenamiento al set de prueba. Los métodos seleccionados para hacer la evaluación son los siguientes:
* __kNN__
* __Gaussian Naive Bayes__
* __Decision Tree__
* __Kernel SVM__
* __Passive Aggressive Classifier__
* __Extremely Randomized Trees__ (Método de Ensamble)

### Tasa de Aprendizaje

Se ha tomado la decisión de utilizar una función denominada _plot__learning__curve_ para visualizar una gráfica que provee información sobre la tasa de aprendizaje obtenida en cada método de Aprendizaje Supervisado. Se ha optado por evaluar cada método mediante la función _learning__curve()_, la cual determina los resultados de las pruebas de entrenamiento y validación cruzada para diferentes tamaños de conjuntos de entrenamiento.

```python
# Bibliotecas requeridas para utilizar los clasificadores
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.model_selection import learning_curve
from sklearn.metrics import accuracy_score
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import OrthogonalMatchingPursuit
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier


def plot_learning_curve(estimator, title, X, Y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):

    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, Y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

Definición de los parámetros de entrada de los clasificadores:

```python
training_vector = trn_set_df.values
target = trn_tgt_df 
testing_vector = tst_set_df.values
testout = tst_tgt_df
ylim = (0.1, 1.05)
```

### 1. kNN

Para este método se realizaron variaciones en el valor de _k_ (cantidad de vecinos cercanos) y se observó que al incrementarlos, la precisión incrementó como se esperaba.

```python
title = "Learning Curves: kNN"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)

# Clasificador kNN con k=5
estimator = KNeighborsClassifier(n_neighbors=7)

plot_learning_curve(estimator, title, training_vector, target, cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = KNeighborsClassifier(n_neighbors=9) 
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
kNN_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', kNN_result)
```

### 2. Gaussian Naive Bayes

En este caso se varió el parámetro _var__smoothing_, que define una parte de la mayor varianza de todas las características y que se agrega a las demás varianzas para dar estabilidad al cálculo. Se identificó que al incrementar este parámetro, la precisión mejoró.

```python
title = "Learning Curves: Gaussian Naive Bayes"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)

# Clasificador GaussianNB
estimator = OneVsRestClassifier(GaussianNB(var_smoothing = 1e-09))

plot_learning_curve(estimator, title, training_vector, target, ylim, cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = GaussianNB(var_smoothing = 1e-09)
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
GNB_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', GNB_result)
```

### 3. Decision Tree

Para este método se varía el parámetro denominado como _max__depth_ que determina la máxima profundidad del árbol de decisión. Se identificó que al disminuir dicha profundidad, se obtiene una mejora en la precisión como se esperaba.

```python
title = "Learning Curves: Decision Tree"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)

# Clasificador DecisionTree con profundidad = 5
estimator = DecisionTreeClassifier(max_depth=5)

plot_learning_curve(estimator, title, training_vector, target, ylim, cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = DecisionTreeClassifier(max_depth=4)
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
DT_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', DT_result)
```

### 4. Kernel SVM

Para el Kernel SVM se utilizó el kernel por defecto denominado RBF (Radius Basis Function), variando su propiedad _gamma_ que define hasta qué punto llega la influencia de una sola muestra del set de entrenamiento. Con valores bajos significa 'lejos' y con valores altos significa 'cerca'. En este caso se obtuvo mejor resultado al variar la magnitud de dicho parámetro.

```python
title = "Learning Curves: SVM con Kernel RBF"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)

# Clasificador Kernel SVM - RBF con gamma=0.001
estimator = OneVsRestClassifier(SVC(kernel='rbf', gamma=0.001))

plot_learning_curve(estimator, title, training_vector, target, (0.7, 1.01), cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = SVC(kernel='rbf', gamma=0.001) 
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
SVM_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', SVM_result)
```

### 5. Passive Aggressive Classifier

Este método fue utilizado variando el número máximo de iteraciones sobre los datos de entrenamiento (_max_iter_) y la tolerancia aceptada. El incremento de la cantidad de iteraciones pierde sentido cuando ya se alcanzó la tolerancia definida.

```python
title = "Learning Curves: Passive Aggressive Classifier"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=10)

# Clasificador PassiveAggressiveClassifier
estimator = OneVsRestClassifier(PassiveAggressiveClassifier(tol=0.0001, max_iter=20))

plot_learning_curve(estimator, title, training_vector, target, ylim, cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = PassiveAggressiveClassifier(tol=0.0001, max_iter=25)
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
PAC_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', PAC_result)
```

## Métodos de Ensamble

Los métodos de ensamble son algoritmos de aprendizaje que construyen un conjunto de clasificadores para aplicarlos sobre nuevos puntos de datos, mediante la toma de un voto (ponderado) de sus predicciones. El objetivo de estos métodos está dado por la combinación de las predicciones de varios estimadores base construidos con un algoritmo de aprendizaje dado, de manera que sea posible mejorar la generalización y robustez en un único estimador [1].

En general, los métodos de ensamble se dividen en las siguientes familias:

* __Métodos de Promediación (Averaging Methods):__ el principio consiste en construir varios estimadores de forma independiente y luego promediar sus predicciones. En promedio, el estimador combinado suele ser mejor que cualquiera de los estimadores de base única porque se reduce la varianza.

  * __Ejemplos:__ 
    * Bagging methods
    * Forests of randomized trees

* __Métodos de Impulso (Boosting Methods):__ en estos métodos los estimadores base se construyen secuencialmente y se intenta reducir el sesgo del estimador combinado. La motivación es combinar varios modelos débiles para producir un conjunto poderoso.

  * __Ejemplos:__ 
    * AdaBoost
    * Gradient Tree Boosting

### 6. Extremely Randomized Trees

En este caso se utiliza un método de promediación llamado Extremely Randomized Trees. Al igual que en el Random Forest tradicional, se utiliza un subconjunto aleatorio de características candidatas, pero en lugar de buscar los umbrales más discriminativos, los umbrales se trazan al azar para cada característica candidata y el mejor de estos umbrales generados aleatoriamente se elige como la regla de división. Esto generalmente permite reducir más la variación del modelo; pero a costa de un aumento ligeramente mayor en el sesgo [2].

Se identificó que incrementando la cantidad de estimadores (_n__estimators_) mejora levemente la precisión. Este parámetro equivale a la cantidad de árboles en el bosque. Otro parámetro que se varió fue el _min__samples__split_, el cual mejoró la precisión al incrementar su magnitud. Nótese que el valor del parámetro _max__depth_ se ha dejado como _None_, en cuyo caso los nodos de los árboles se expanden hasta que todas las hojas contengan menos muestras que las indicadas por _min__samples__split.

```python
title = "Learning Curves: Extremely Randomized Trees"

# Validacion cruzada con 100 iteraciones y 20% de datos seleccionados al azar
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=10)

# Clasificador PassiveAggressiveClassifier
estimator = OneVsRestClassifier(ExtraTreesClassifier(n_estimators=10, max_depth=None, min_samples_split=2, random_state=0))

plot_learning_curve(estimator, title, training_vector, target, ylim, cv=cv, n_jobs=4)
plt.show()

# Utilizando Set de Prueba 
classifier = ExtraTreesClassifier(n_estimators=20, max_depth=None, min_samples_split=10, random_state=0) 
classifier = classifier.fit(training_vector, target)
predicted = classifier.predict(testing_vector)

# Printing the results
ERT_result = accuracy_score(testout, predicted)
print ('Accuracy Score :', ERT_result)
```
