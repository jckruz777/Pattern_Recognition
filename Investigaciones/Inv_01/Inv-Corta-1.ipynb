## Tutorial de PANDAS para Pre-Procesamiento de Datos



#### Se importa la Biblioteca de PANDAS y NumPy


```python
import pandas as pd
import numpy as np
```

#### Se importan los datos mediante la lectura de un archivo  _iris.csv_ modificado a partir del original _iris-data.csv_. En este caso se utilizará información sobre el iris de las flores. El set utilizado se puede obtener del siguiente enlace: [Datos de Iris](https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv) y ha sido modificado brevemente mediante la introducción de valores no conocidos NaN en algunas celdas.

```python
df = pd.read_csv("iris.csv")
df
```

### Replace
#### Convertir valores desconocidos en NaN

```python
df = df.replace('?', np.nan)
df
```

### Dropna
#### El comando _dropna_ se utiliza para no considerar celdas donde se tienen valores _np.nan_ (desconocidos). Los parámetros que el comando recibe definen las celdas sobre las que éste se debe aplicar.

* __axis:__ puede referirse a filas (__"rows"__ | 0) ó a columnas (__"columns"__ | 1). Si no se indica, por defecto se tomarán las filas.
* __how:__ puede referirse a los casos donde todas las celdas son desconocidas (__"all"__) ó donde al menos una de las celdas es desconocida (__"any"__). Si no se indica, el valor por defecto será __"any"__.
* __thresh:__ indica el número máximo de celdas desconocidas requerido para mantener una fila.
* __subset:__ utilizado para indicar el nombre de las columnas en las cuales verificar celdas desconocidas. Estos nombres deben ir con comillas simples separados por coma dentro de paréntesis cuadrados. Por ejemplo: _subset=['age', 'classification']._
* __inplace:__ puede usarse para mantener (__True__) el marco de datos con entradas válidas en la misma variable.

#### Por ejemplo, en este caso se desea desechar las filas (axis=0) donde al menos una celda es desconocida (how="any"):

```python
df_dropna = df.dropna(axis=0, how="any")
df_dropna
```

### Isna
#### El comando _isna_ puede ser utilizado para identificar mediante valores booleanos las celdas cuyo valor es desconocido (True)

```python
df_isna = df.isna()
df_isna
```

### Notna
#### El comando _notna_ puede ser utilizado para identificar mediante valores booleanos las celdas cuyo valor es conocido (True)

```python
df_notna = df.notna()
df_notna
```

### Fillna
#### El comando _fillna_ es utilizado para asignar valores a las celdas desconocidas. La asignación se realiza de acuerdo a lo indicado por los parámetros.

* __value:__ representa el valor a usar para rellenar las celdas con valor desconocido. También se puede utilizar un diccionario (__dict__), una serie (__Series__) ó un marco de datos (__DataFrame__) con valores que especifican los valores a utilizar por cada índice (para series) ó columna (para marcos de datos). Los valores que no se encuentren especificados en el diccionario, serie ó marco de datos utilizado como parámetro, no serán asignados. No se pueden usar listas como parámetros. Por ejemplo: se puede definir la siguiente variable _values = {'A': 0, 'B': 1, 'C': 2, 'D': 3}_ y aplicarlo con el comando _fillna(value=values)_. Esto reemplazará los valores desconocidos en la columna etiquetada como 'A' por ceros, en la columna etiquetada como 'B' por unos y así sucesivamente.

* __method:__ define el método a utilizar para rellenar las celdas desconocidas en una serie reindexada. Al utilizar este parámetro se propaga el último valor válido hacia adelante (__'pad'__ | __'ffil'__) ó el siguiente valor válido hacia atrás (__'backfill'__ | __'bfill'__) para rellenar las celdas desconocidas. Por defecto el parámetro posee el valor None.

* __axis:__ permite definir si el comando debe aplicarse por índice (0 | 'index') ó por columna (1 | 'column').

* __inplace:__ utilizado para indicar mediante un booleano si el rellenado debe hacerse en orden o no. Su valor es __False__ por defecto. 
* __limit:__ si se utiliza _method_, este parámetro indica el máximo número de celdas desconocidas NaN consecutivas a rellenar hacia adelante o hacia atrás. Debe ser mayor a cero si no es None.
* __downcast:__ corresponde a un diccionario (dict) de tipos (item->dtype) que indica los tipos de datos con los que se desea realizar el relleno. También se puede utilizar el valor __'infer'__ del parámetro para tratar de realizar el cast de tipos (de _float64_ a _int64_ por ejemplo) en caso de ser posible.  

```python
df_fillna = df.fillna(method='ffill')
df_fillna
```

### Interpolate

#### El comando _interpolate_ se utiliza para interpolar datos por diferentes métodos. El tipo de interpolación se define a través de varios parámetros.

* __method:__ indica la técnica de interpolación a utilizar. Puede ser:
 - _'linear':_ ignora el índice y considera que los valores están separados por distancias iguales. Este es la única técnica soportada cuando se utilizan índices múltiples. 
 - _'time':_ se debe asignar un valor correspondiente al tamaño de intérvalo que trabaja sobre datos de mayor resolución para interpolar. 
 - _'index', 'values':_ utiliza los valores numéricos actuales de índices.  
 - _'pad':_ rellena las celdas desconocidas NaN utilizando valores existentes.
 - _'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'spline', 'barycentric', 'polynomial':_ algunos de estos parámetros requieren de la indicación del orden (__order__) de la interpolación. Por ejemplo: _interpolate(method='polynomial', order=4)_. En este caso se utilizan valores numéricos para los índices.

* __axis:__ indica el eje sobre el cual se aplicará la interpolación. Puede ser por índice (0 | __'índex'__) ó por columnas (1 | __'columns'__). El valor predeterminado del atributo es __None__.

* __limit:__ indica el número máximo de celdas NaN consecutivas a rellenar. Dicho número debe ser mayor que cero.

* __inplace:__ actualiza o no, mediante un booleano, los datos en orden si es posible. Por defecto el valor de este atributo es __False__.

* __limit_direction:__ se rellenan las celdas NaN en la dirección indicada, la cual puede ser hacia adelante (__'forward'__), hacia atrás (__'backward'__) ó ambas (__'both'__). Por defecto el valor de este atributo es __forward__.

* __limit_area:__ puede usarse para rellenar celdas NaN consecutivas con la restricción definida por el límite. Puede indicarse que no se establezca restricción (__None__), que solamente las celdas NaN rodeadas por valores válidos sean rellenadas (__'inside'__) ó que solamente las celdas NaN que se encuentran afuera de los valores válidos conocidos (__'outside'__), en cuyo caso se aplicaría una extrapolación.   

* __downcast:__ trata de realizar el cast si es posible. Puede indicarse como inferencia (__'infer'__) ó no realizarse (__None__). Éste último es el valor por defecto del atributo.

* __kwargs:__ permite especificar valores mediante el teclado para considerarlos en la función de interpolación.

#### Por ejemplo, en el siguiente caso se rellenan las celdas desconocidas NaN de la columna __'sepal_length'__ con valores existentes.

```python
df_interpolate = df['sepal_length'].interpolate(method='pad', limit=2)
df_interpolate
```

## Análisis Visual de Datos con Matplotlib

#### Se importa la biblioteca de gráficos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Gráfico de Dispersión:
#### Utilizado para visualizar la relación entre dos variables mediante la observación de la distribución de los datos.

```python
# Cargar los datos de iris de flores
from sklearn.datasets import load_iris

# Extrayendo los datos y las clases
feats = load_iris()['data']
target = load_iris()['target']

# Creando objeto para las gráficas
_, ax = plt.subplots()

# Asignando colores por clase
colors = ("red", "blue", "green")
groups = ("iris setosa", "iris versicolor", "iris virginica")

# Iterando por cada clase
for i in np.unique(target):
    mask = target == i
    plt.scatter(feats[mask, 0], feats[mask, 1], color=colors[i], label=groups[i])
    
# Asignando títulos a los ejes
ax.set_title("Petal Width vs Sepal Length")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Petal Width")

# Agregando las etiquetas de los grupos
ax.legend()

# Mostrando la gráfica
plt.show()
```

## Histograma

#### Utilizado para analizar la frecuencia con que aparecen los valores que poseen los atributos en un set de datos.

```python
import matplotlib.pyplot as plt
from sklearn import datasets

# Otra forma de cargar los datos de iris de flores
iris= datasets.load_iris()

# Establecimiento del grid para mostrar varios gráficos
fig, axes = plt.subplots(nrows= 2, ncols=2, figsize=(10,10))

# Asignación de colores
colors= ['blue', 'red', 'green']

# Iteración por filas pertenecientes a cada clase
for i, ax in enumerate(axes.flat):
    for label, color in zip(range(len(iris.target_names)), colors):
        ax.hist(iris.data[iris.target==label, i],
                label= "iris " + iris.target_names[label], color=color)
        ax.set_xlabel(iris.feature_names[i])  
        ax.legend(loc='upper right')


plt.show()
```

### Matriz de Correlación
#### Utilizada para analizar la interdependencia en relaciones asociadas entre parejas de variables, expresada como valores en el rango [0,1]. La diagonal principal está dada por unos debido a que se trata de la correlación entre la misma variable (100%). 

#### NOTA: Para mapas de calor se requiere de la biblioteca _seaborn_.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Se cargan los datos de iris de flores a partir de un archivo
iris= pd.read_csv("iris-data.csv")

# Se obtienen los valores de correlación entre las variables
corr = iris.corr()

# Se define el estilo y los parámetros de la matriz de correlación a mostrar gráficamente
plt.figure(figsize=(10,8)) 
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           cmap='viridis', annot=True)
plt.show()
```

### Diagrama de Caja
#### Utilizado para representar de forma gráfica conjuntos de datos numéricos a través de sus cuartiles (datos que componen las cuartas partes del set ubicados en los rangos dados por 25%, 50%, 75% y el 100%). El diagrama de caja muestra la mediana y los cuartiles de los datos, representando además los valores atípicos.

```python
import pandas as pd
import seaborn as sns

# Se cargan los datos de iris de flores a partir de un archivo
iris= pd.read_csv("iris-data.csv")

ax= sns.boxplot(x="species", y="petal_length", data=iris)
ax= sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="gray")
plt.show()
```
